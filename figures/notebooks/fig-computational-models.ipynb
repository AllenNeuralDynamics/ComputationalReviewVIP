# Figure 12.1 — Computational models of VIP circuit function

Reproduces 9 audited panels (status=CAVEAT) from the slim figure pack.

```python
import json, matplotlib.pyplot as plt
from collections import Counter
data = json.load(open('/sessions/operon-k-b789538d-548a-3a1a/mnt/artifacts/proj_51704bdb3ee3/400fe0d3-06c9-4c25-9078-b3ea06b474a2/vd73f4b02_sec12_figure_pack_slim.json'))
ap = data['figures'][0]['audited_panels']
print('panels:', len(ap))
```

```python
def cnt(entries, field):
    return Counter(e[field] for e in entries if e.get(field))
def filter_cat(entries, cat):
    return [(e['value'], e['count']) for e in entries if e['category'] == cat]
p2_mt = cnt(ap[2]['figure_data']['entries'], 'model_type')
p4_mt = cnt(ap[4]['figure_data']['entries'], 'model_type')
p6_mt = cnt(ap[6]['figure_data']['entries'], 'model_type')
p3_vr = filter_cat(ap[3]['figure_data']['entries'], 'VIP_role')
p5_vr = filter_cat(ap[5]['figure_data']['entries'], 'VIP_role')
p7_vr = filter_cat(ap[7]['figure_data']['entries'], 'vip_role')
p8_vr = [(e['value'], e['count']) for e in ap[8]['figure_data']['entries']]
```

```python
def hbar(ax, items, title, color):
    items = sorted(items, key=lambda x: x[1])
    labels = [k for k,_ in items]; vals = [v for _,v in items]
    bars = ax.barh(labels, vals, color=color, edgecolor='black', linewidth=0.4)
    ax.set_title(title, fontsize=9, pad=4)
    ax.tick_params(axis='both', labelsize=7)
    ax.spines[['top','right']].set_visible(False)
    for b, v in zip(bars, vals):
        ax.text(v+max(vals)*0.02, b.get_y()+b.get_height()/2, str(v), va='center', fontsize=7)
    ax.set_xlim(0, max(vals)*1.18)

fig, axes = plt.subplots(3, 3, figsize=(13, 11))
hbar(axes[0,0], list(Counter(e['vip_inclusion'] for e in ap[0]['figure_data']['entries']).items()),
     'A. VIP-included audit (per-row)\n25 classifiable / 40 total', '#4C72B0')
hbar(axes[0,1], [(e['bucket_label'], e['count']) for e in ap[1]['figure_data']['entries']],
     'B. Aggregate counts\n(VIP_included vs VIP_excluded; n=25)', '#55A868')
hbar(axes[0,2], list(p2_mt.items()), 'C. Round-2 taxonomy: model_type (n=40)', '#C44E52')
hbar(axes[1,0], p3_vr, 'D. Round-2 distribution: VIP_role (n=40)', '#8172B2')
hbar(axes[1,1], list(p4_mt.items()), 'E. Round-3 taxonomy: model_type (n=40)', '#C44E52')
hbar(axes[1,2], p5_vr, 'F. Round-3 distribution: VIP_role (n=40)', '#8172B2')
hbar(axes[2,0], list(p6_mt.items()), 'G. Round-4 taxonomy: model_type (n=40)', '#C44E52')
hbar(axes[2,1], p7_vr, 'H. Round-4 distribution: vip_role (n=40)', '#8172B2')
hbar(axes[2,2], p8_vr, 'I. Coherent VIP roles (cumulative; n=160 papers, count>=3)', '#CCB974')
fig.suptitle('Figure 12.1 - Computational models of VIP circuit function (audited inventory)', fontsize=12, y=0.995)
fig.tight_layout(rect=(0, 0, 1, 0.97))
fig.savefig('fig-computational-models.png', dpi=200, bbox_inches='tight')
fig.savefig('fig-computational-models.pdf', bbox_inches='tight')
```

## Figure 12.1 — Computational models of VIP circuit function (audited inventory)

**Caption.** Audited inventory of computational/modeling studies of VIP circuit function across four curation rounds. Panel A renders the per-row VIP-inclusion classification from comparison `cluster_11_computational_models__0` (25 of 40 papers classifiable into VIP_included = 21 vs VIP_excluded = 4; 15 unclassifiable rows dropped). Panel B renders the matching aggregate-count panel `__1` (same 21 vs 4 split, n = 25 classifiable models). Panels C, E, G show model-type breakdowns of the round-2 / round-3 / round-4 architectural taxonomies (n = 40 each round). Panels D, F, H show the corresponding VIP-role distributions for each round. Panel I shows the cumulative coherent VIP-role categories surviving the round-4 restructure (count ≥ 3 over n = 160 papers). All nine panels were audited with status = CAVEAT; the verbatim per-panel caveats are reproduced below and must travel with the figure.

**Panel A caveat:** Restructure drops the 15 unclassifiable rows and claims to retain 25 rows split into 21 VIP-included and 4 VIP-excluded — the entry count (25) matches that arithmetic. But the entries themselves carry only cite_key/doi/study_label; metric_label, value, n_definition, scope, and value_source_sentence are all None/empty. The per-row VIP-inclusion classification that is the entire substantive content of the panel is not encoded in the entries, so it is not auditable from the packet — only the restructure_notes claim is available. As a bare list of papers it's homogeneous (computational microcircuit models) and study_labels are first-author surnames matching the listed cite_keys, but the categorical inventory cannot be verified row-by-row. Conservative CAVEAT: needs an explicit per-row VIP_included/VIP_excluded field to be auditable.

**Panel B caveat:** Aggregate-count panel is supposed to report stratified counts (claimed 21 VIP-included vs 4 VIP-excluded, per restructure_notes), but the two entries have value='VIP_included' and value='VIP_excluded' — these are bucket labels, not counts. There is no numeric count field anywhere in the entries, so the panel as encoded cannot be rendered as a count. cite_key, doi, study_label are all None (acceptable for an aggregate row), but without the actual count values this is a structural defect: the data does not contain the metric the metric_label describes. The restructure also removes the model_type aggregate but the replacement count panel is not actually populated. Needs the entries to carry numeric counts (e.g., value=21 / value=4) plus a derivation note pointing to the __0 inventory.

**Panel C caveat:** Taxonomy spans heterogeneous brain regions (neocortex V1/PFC/A1/parietal/L1, hippocampus CA1/CA3, BLA, olfactory cortex); category counts are not region-controlled. | model_type categories are non-orthogonal (biophysical = implementation level; rate/spiking/neural_mass = formalism; predictive_coding = paradigm); papers may legitimately fit more than one bucket. | VIP_role labels mix mechanism (disinhibition_gating) with function (prediction_error, gain_modulation); no source sentence is shown to verify each assignment. | Per-paper model scale varies from single-cell biophysical models to multi-million-neuron mesocircuits; each contributes equal weight. | interneuron_classes column shows 8 distinct compositions; counts do not separate 2-class from 4-class models. | 3 of 40 DOIs lack proper author-year citation keys (Unmapped_* placeholders): 10.1073/pnas.2408966122 and 10.7554/elife.100268.3.

**Panel D caveat:** Counts derive from the round-2 taxonomy (comparison __2) and inherit its non-orthogonal categories. | model_type buckets mix formalism (rate/spiking/neural_mass), implementation (biophysical), and paradigm (predictive_coding); a single paper can legitimately fit several buckets. | VIP_role labels mix mechanism and function; the 60% 'disinhibition_gating' fraction is partly a labeling-level artifact. | Distribution does not control for brain region, model scale, or interneuron-class composition; counts are unweighted. | n=40 papers across all three category axes.

**Panel E caveat:** Sample of 40 modeling papers spans heterogeneous brain regions and model scales (single-cell to whole-brain). | model_type schema differs from round-2 (adds mean_field, anatomical_basis; drops reservoir_computing, neural_mass) — distributions are not directly comparable to round-2. | VIP_role schema adds 4 new categories vs round-2; categories overlap semantically (attention/top-down ≈ context_modulation ≈ gain_modulation in some entries). | interneuron_classes vary from VIP-only (5 papers) to 4-class compositions; counts do not separate by composition. | Citation 10.48550/arxiv.1801.00062 is an Unmapped_* placeholder and lacks a proper author-year cite_key.

**Panel F caveat:** Counts derive from the round-3 taxonomy (comparison __4) and inherit its non-orthogonal categories. | Round-3 schema is NOT identical to round-2 schema: VIP_role has 9 categories (vs 5), model_type adds mean_field and anatomical_basis. Round-2 vs round-3 distributions cannot be directly compared. | Round-2 and round-3 paper samples are disjoint (n=40 each, 0 overlap). | Counts are unweighted across model scale (single-cell to whole-brain) and brain region. | interneuron_classes counts do not separate by class breadth (VIP-only vs 4-class compositions).

**Panel G caveat:** Table is a taxonomic survey of computational/modeling studies, not a comparison of measured values; rows differ in brain region, species (incl. songbird), modeling formalism, and cell-type composition. | model_type and vip_role labels are non-orthogonal categories assigned by the curators; some entries lack VIP cells but still receive a vip_role tag. | Entry cluster_11_finding_123 cites DOI 10.64898/2025.12.21.695847, which is unmapped in the citation_key_map (cite_key=Unmapped_3be007); abstract resolves to Yao et al. 2025.

**Panel H caveat:** Counts are over the same 40 round-4 architectural-taxonomy entries; categories are not mutually exclusive and were assigned by the curators. | vip_role counts include studies that do not include VIP interneurons in the modeled population.

**Panel I caveat:** Restructure correctly isolates vip_role rows (single category, single n=160 denominator, count>=3 cutoff applied), dropping model_type and the anatomical_basis tag — internal homogeneity on metric_definition is satisfied. | However, role labels are free-text modeler-assigned tags drawn from a cumulative pool of n=160 papers across four rounds, and the restructure_notes acknowledges 'some papers contribute to multiple categories' — bars are not mutually exclusive and cannot be interpreted as a proportion partition. | This is a literature-orientation count, not an audit-extracted primary measurement (homogeneity_check explicitly flags this); the figure should ship with the caveat already specified in restructure_notes. | Conservative CAVEAT: structurally clean, but inherent free-text/non-exclusive caveats must accompany the panel.
